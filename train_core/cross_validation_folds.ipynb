```python
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold
import pandas as pd
```

```python
df = pd.read_csv('/Users/teostock/PycharmProjects/DataScience/data/train_targets_scored.csv')
df.loc[:, "kfold"] = -1
df = df.sample(frac = 1).reset_index(drop = True)
targets = df.drop('sig_id', axis = 1).values
```

```python
mskf = MultilabelStratifiedKFold(n_splits = 5)
for fold_, (trn_, val_) in enumerate(mskf.split(X = df, y = targets)):
    df.loc[val_, "kfold"] = fold_
```

```python
df.to_csv('/Users/teostock/PycharmProjects/DataScience/data/MultilabelStratifiedKFold.csv', index = False)
df
```
